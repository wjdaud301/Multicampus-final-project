```python
from pyspark.conf import SparkConf
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType,StructField, StringType, IntegerType
from pyspark.sql.functions import array_contains, udf

from pymongo import MongoClient
import pandas as pd
```

```python
spark_A = SparkSession \
    .builder \
    .appName("corona") \
    .config("spark.mongodb.input.uri", "mongodb://localhost:27017") \
    .config("spark.mongodb.input.database","ojo_db") \
    .config("spark.mongodb.input.collection", "coronaAPI") \
    .config("packages org.mongodb.spark:mongo-spark-connector_2.12:3.0.1") \
    .getOrCreate()

sc_A =spark_A.sparkContext

spark_A


df_API = spark_A.read.format("com.mongodb.spark.sql.DefaultSource").load()
df_API.show()
spark_A.stop()
```
