```python
import re

def extract(x):
    p = re.compile('\"[가-힣\t\n\r\f\v\s\_가-힣]+\"')
    m = p.findall(x)
    return m[0]

info_pd.content.apply(lambda x : extract(x))
```

```python
def tokenize(x):
    p = re.compile('[가-힣]+')
    m = p.findall(x)
    return ' '.join(m)

info_pd.content.apply(lambda x : tokenize(x))
```
